# Q2 = Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes?
- Ans ▶
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BEM Example</title>
    <link rel="stylesheet" href="styles.css">
    <style>
.content {
    padding: 20px;
    background-color: #f0f0f0;
    border: 1px solid #ccc;
}
.first {
    color: blue;
    font-weight: bold;
}
.second {
    color: green;
}

.third {
    color: orange;
    font-style: italic;
}

.fourth {
    background-color: yellow;
    padding: 5px;
    border-radius: 3px;
}
    </style>
</head>
<body>

    <div class="content">
        <p class="first">This is the primary paragraph.</p>
        <p class="second">This is the secondary paragraph.</p>
        <p class="third">This is the tertiary paragraph.</p>
        <p class="fourth">This is the highlighted paragraph.</p>
    </div>

</body>
</html>
